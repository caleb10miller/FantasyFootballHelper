```python
# PPR by team position group
threshold = 50
ff_filtered = ff_data[ff_data['PPR Fantasy Points Scored'] >= threshold]
team_performance = ff_filtered.groupby(['Team', 'Position'])['PPR Fantasy Points Scored'].mean().unstack()

plt.figure(figsize=(12, 6))
ax = sns.heatmap(team_performance, cmap="coolwarm", annot=True, fmt=".1f", linewidths=0.5)

# custom labels
ax.set_yticks(range(len(team_performance))) 
ax.set_yticklabels(team_performance.index, rotation=0, ha="right") 

plt.title(f"Average PPR Fantasy Points by Team Position Group (Min {threshold} PPR Points)")
plt.xlabel("Position")
plt.ylabel("Team")
plt.show()
```

```python
# ADP Gap and first attempt at clustering
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

sleeper_data = ff_data[['Player Name', 'Position', 'Average ADP', 'PPR Fantasy Points Scored', 'Games Played']].dropna()
sleeper_data['Expected Rank'] = sleeper_data['Average ADP'].rank(method="min")
sleeper_data['Actual Rank'] = sleeper_data['PPR Fantasy Points Scored'].rank(ascending=False, method="min")
sleeper_data['ADP Gap'] = sleeper_data['Expected Rank'] - sleeper_data['Actual Rank']

scaler = StandardScaler()
X = scaler.fit_transform(sleeper_data[['ADP Gap', 'PPR Fantasy Points Scored']])
```

```python
# elbow method to find optimal k
inertia = []
K_range = range(2, 10)
for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K_range, inertia, marker='o', linestyle='--', label="Inertia")
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.title('Optimal K')
plt.legend()
plt.show()
```

```python
optimal_k = 4
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
sleeper_data['Cluster'] = kmeans.fit_predict(X)
sleeper_data['Cluster'] = sleeper_data['Cluster'].astype(str)

plt.figure(figsize=(10, 6))
sns.scatterplot(
    x=sleeper_data['ADP Gap'], 
    y=sleeper_data['PPR Fantasy Points Scored'], 
    hue=sleeper_data['Cluster'], 
    palette="coolwarm", 
    alpha=0.8
)
plt.axvline(0, color='gray', linestyle='--', alpha=0.7) 
plt.xlabel("ADP Gap (Positive = Sleeper, Negative = Bust)")
plt.ylabel("PPR Points Scored")
plt.title("Clustering Sleepers and Busts")
plt.legend(title="Cluster")
plt.show()
```

```python
# top sleepers
sleepers = sleeper_data[sleeper_data['ADP Gap'] > 0].sort_values(by='ADP Gap', ascending=False)
print("Top Sleepers:")
print(sleepers[['Player Name', 'Position', 'Average ADP', 'PPR Fantasy Points Scored', 'ADP Gap']].head(10))
```

```python
# biggest busts
busts = sleeper_data[sleeper_data['ADP Gap'] < 0].sort_values(by='ADP Gap', ascending=True)
print("Biggest Busts:")
print(busts[['Player Name', 'Position', 'Average ADP', 'PPR Fantasy Points Scored', 'ADP Gap']].head(10))
```

```python
# top sleepers by position
top_sleepers_by_position = sleepers.groupby('Position').apply(lambda x: x.sort_values(by='ADP Gap', ascending=False).head(5))
top_sleepers_by_position = top_sleepers_by_position.reset_index(drop=True)

print("Top Sleepers by Position:")
print(top_sleepers_by_position[['Player Name', 'Position', 'Average ADP', 'PPR Fantasy Points Scored']])
```

```python
# biggest busts by position
biggest_busts_by_position = busts.groupby('Position').apply(lambda x: x.sort_values(by='ADP Gap').head(5))
biggest_busts_by_position = biggest_busts_by_position.reset_index(drop=True)

print("Biggest Busts by Position:")
print(biggest_busts_by_position[['Player Name', 'Position', 'Average ADP', 'PPR Fantasy Points Scored']])
```
