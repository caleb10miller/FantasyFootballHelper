```python
# ADP Gap and second attempt at clustering, accounting for position
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

sleeper_data = ff_data[['Player Name', '2024 Position', '2024 Average ADP', '2024 PPR Fantasy Points Scored']].dropna()

sleeper_data['2024 Expected Rank'] = sleeper_data.groupby('2024 Position')['2024 Average ADP'].rank(method="min")
sleeper_data['2024 Actual Rank'] = sleeper_data.groupby('2024 Position')['2024 PPR Fantasy Points Scored'].rank(ascending=False, method="min")
sleeper_data['2024 ADP Gap'] = sleeper_data['2024 Expected Rank'] - sleeper_data['2024 Actual Rank']

scaler = StandardScaler()
X = scaler.fit_transform(sleeper_data[['2024 ADP Gap', '2024 PPR Fantasy Points Scored']])
```

```python
optimal_k = 4  
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
sleeper_data['Cluster'] = kmeans.fit_predict(X)
sleeper_data['Cluster'] = sleeper_data['Cluster'].astype(str)

plt.figure(figsize=(10, 6))
sns.scatterplot(
    x=sleeper_data['2024 ADP Gap'], 
    y=sleeper_data['2024 PPR Fantasy Points Scored'], 
    hue=sleeper_data['Cluster'], 
    palette="coolwarm", 
    alpha=0.8
)
plt.axvline(0, color='gray', linestyle='--', alpha=0.7)
plt.xlabel("2024 ADP Gap (Positive = Sleeper, Negative = Bust)")
plt.ylabel("2024 PPR Fantasy Points Scored")
plt.title("Position-Adjusted Clustering: Sleepers & Busts")
plt.legend(title="Cluster")
plt.show()
```

```python
# find cluster centers to pull representative player
cluster_centers = scaler.inverse_transform(kmeans.cluster_centers_)

representative_players = []
for cluster_idx, center in enumerate(cluster_centers):
    sleeper_data["distance"] = np.linalg.norm(X - center, axis=1)
    closest_player = sleeper_data[sleeper_data["Cluster"] == str(cluster_idx)].nsmallest(5, "distance")
    representative_players.append(closest_player)

representative_players_df = pd.concat(representative_players)
print(representative_players_df[['Player Name', 'Cluster', '2024 ADP Gap', '2024 PPR Fantasy Points Scored']])
```

```python
# another attempt at identifying surprising performers
# drop rookies
ff_data = ff_data.dropna(subset=["2022 PPR Fantasy Points Scored", "2023 PPR Fantasy Points Scored"])

# calculate fantasy point gains
ff_data["FP Growth 2022-2023"] = ff_data["2023 PPR Fantasy Points Scored"] - ff_data["2022 PPR Fantasy Points Scored"]
ff_data["FP Growth 2023-2024"] = ff_data["2024 PPR Fantasy Points Scored"] - ff_data["2023 PPR Fantasy Points Scored"]

# breakout candidates by biggest year-over-year improvement
breakouts = ff_data[(ff_data["FP Growth 2023-2024"] > 50) & (ff_data["2024 Age"] < 32)]
breakouts = breakouts[["Player Name", "2024 Position", "2024 Age", "FP Growth 2023-2024", "2024 PPR Fantasy Points Scored"]]
breakouts = breakouts.sort_values(by="FP Growth 2023-2024", ascending=False)
print("Top Breakout Candidates (Excluding Rookies):")
print(breakouts)

plt.figure(figsize=(10, 6))
sns.barplot(
    y=breakouts["Player Name"], 
    x=breakouts["FP Growth 2023-2024"], 
    hue=breakouts["2024 Position"],
    palette=position_colors,
    dodge=False
)
plt.xlabel("Fantasy Points Increase (2023 to 2024)")
plt.ylabel("Player Name")
plt.title("Top Breakout Candidates (Excluding Rookies)")
plt.grid(True)
plt.show()
```
